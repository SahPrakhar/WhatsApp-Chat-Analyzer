```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with AI-Training & Placement - Copy.txt','r',encoding='utf-8')
```

```python
data =f.read()
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages= re.split(pattern,data)[1:]
messages
```

```python
dates =re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

#converting message data type
'''
#df['Message_Date'] = pd.to_datetime(df['Message_Date'], format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'Message_Date': 'date'}, inplace=True)
df.head()
'''

df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M - ')
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ', errors='coerce', dayfirst=True)
df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
#sepatate users and messages
users=[]
messages=[]
for message in df['user_message']:
  entry=re.split('([\w\W]+?):\s', message)
  if entry[1:]: #user name
    users.append(entry[1])
    messages.append(" ".join(entry[2:]))
  else:
    users.append('Group_Notification')
    messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
df['year'] = df['date'].dt.year
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
